```python
import pandas as pd
import numpy as np
import boto3
import pickle
```

```python
from sagemaker import get_execution_role
role = get_execution_role()
```

```python
bucket = 'assignment1bucketssm'
my_path1 = 'folder1/file1.txt'
my_path2 = 'folder1/file2.txt'
my_path3 = 'folder1/file3.txt'
```

```python
s3client = boto3.client('s3')
response1 = s3client.get_object(Bucket=bucket, Key=my_path1)
body = response1['Body'].read()

data1 = body.decode('cp1252')
data1
```

```python
s3client = boto3.client('s3')
response2 = s3client.get_object(Bucket=bucket, Key=my_path2)
body2 = response2['Body'].read()

data2 = body2.decode('cp1252')
data2
```

```python
s3client = boto3.client('s3')
response3 = s3client.get_object(Bucket=bucket, Key=my_path3)
body3 = response3['Body'].read()

data3 = body3.decode('cp1252')
data3
```

```python
data = data1+data2+data3
data
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
sent_tokenize(data)
```

```python
words = word_tokenize(data)
words
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
filtered_words = [w for w in words if not w in stopwords.words()]
filtered_words
```

```python
len(words)
```

```python
len(filtered_words)
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
stemmed_result = [stemmer.stem(word) for word in filtered_words]
stemmed_result
```

```python
with open('result.txt', 'w') as my_list_file:
    for element in stemmed_result:
        my_list_file.write('%s\n' % element)
```
